```python
#import libraries
from pathlib import Path
import csv

#import the database in csv format
csvfile_imported = Path("../Pybank/Resources/budget_data.csv")

#Initialize global varibles 
month_summary = []
profit_summary = []
database_summary = []
outliner_summary = []
total_profot_loss = 0
profit_changes = []

#creat a filter using date as identifier to clean the data
month_identifier = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

#open the imported file in READ mode
with open(csvfile_imported, "r") as csvdata:
    
    #pass the csvdata in csv.reader() function with comma as delimiter
    csvreaded = csv.reader(csvdata, delimiter = ",")
    
    #created header for the data 
    #dataheader = next(csvreaded)

    #using identifier to verify whether the data in first coloum is date, if it is, record the date to month_summary and corresponding profit and loss to profit summary. also record the entire row in database_summary. for all the entries that first coloum is not a date, record it in outliner_summary for future inspection.
    for row in csvreaded:
        for x in month_identifier:
            if x in row[0]:
                month_summary.append(row[0])
                profit_summary.append(row[1])
                database_summary.append(row)
            else:
                outliner_summary.append(row)

    ###            
    #print (month_summary)
    #print (profit_summary)
    #print (database_summary)
    #print (outliner_summary)
    ###

    #calculate net total amount of profit and loss over the entire period
    for num in profit_summary:
        total_profot_loss += int(num)
    
    #calculate average changes
    average = round(total_profot_loss / len(profit_summary), 2)
    
    #print (total_profot_loss)
    #print (average)
    
    #calculate the changes between each month and record in profit_changes
    for i in range(len(profit_summary)):
        changes = int(profit_summary[i]) - int(profit_summary[i-1])
        profit_changes.append(changes)
   
    #print(profit_changes)
    #print(len(profit_changes))

    #Initialize variable for max and min calculation 
    maximum_change = 0
    minimum_change = 0
    minimum_date = ""
    maximum_date = ""
    #index is used to count for the corrsponding element in month_summary
    index = 0
    #try to find the max and min and corrsponding date in month_summary
    for i in profit_changes:
        if minimum_change == 0:
            minimum_change = i
            minimum_date = month_summary[index]
        elif i < minimum_change:
            minimum_change = i
            minimum_date = month_summary[index]
        elif i > maximum_change:
            maximum_change = i
            maximum_date = month_summary[index]
        index += 1


    #print (maximum_change)
    #print (maximum_date)
    #print (minimum_change)
    #print (minimum_date)

    #maximum_date_index = month_summary.index(maximum_date)
    #maximum_change_index = profit_changes.index(maximum_change)
    #print(maximum_change_index)
    #print(maximum_date_index)

#out put result to a txt file
output = Path("Output_pybank.txt")
with open(output, "w") as result_file:
    print(f'Financial Analysis', file = result_file)
    print(f'----------------------------', file = result_file)
    print(f'Total Months: {len(month_summary)}', file = result_file)
    print(f'Total: ${total_profot_loss}', file = result_file)
    print(f'Average Changes: ${average}', file = result_file)
    print(f'Greatest Increase in Profits: {maximum_date} (${maximum_change})', file = result_file)
    print(f'Greatest Decrease in Profits: {minimum_date} (${minimum_change})', file = result_file)

1
1
```
